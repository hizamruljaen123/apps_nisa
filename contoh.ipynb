```python
import numpy as np
import matplotlib.pyplot as plt
import librosa

def plot_audio_waveform_amplitude(file1, file2, start_time, end_time):
    """ Plot audio waveform amplitude segments of two audio files in separate graphs """
    # Load audio files
    y1, sr1 = librosa.load(file1, sr=None)
    y2, sr2 = librosa.load(file2, sr=None)

    # Select segment
    start_sample1 = int(start_time * sr1)
    end_sample1 = int(end_time * sr1)
    start_sample2 = int(start_time * sr2)
    end_sample2 = int(end_time * sr2)

    y1_segment = y1[start_sample1:end_sample1]
    y2_segment = y2[start_sample2:end_sample2]

    # Plot the first audio waveform amplitude
    plt.figure(figsize=(14, 5))
    plt.plot(y1_segment, linewidth=1, color='blue')
    plt.title('Reference Audio Amplitude Segment')
    plt.xlabel('Sample Number')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

    # Plot the second audio waveform amplitude
    plt.figure(figsize=(14, 5))
    plt.plot(y2_segment, linewidth=1, color='orange')
    plt.title('Query Audio Amplitude Segment')
    plt.xlabel('Sample Number')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

# Example usage
file1 = 'test.opus'  # Replace with the correct file path
file2 = 'test_1.opus'  # Replace with the correct file path
start_time = 10  # Start time in seconds
end_time = 20    # End time in seconds

plot_audio_waveform_amplitude(file1, file2, start_time, end_time)
```
